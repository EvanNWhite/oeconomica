Worked Example: BA Wage Premium Over Time
Full runnable version — copy each block into a Jupyter cell.
Requires: wages_by_education.csv in the same folder.
(Unlike the skill modules, the column names here are the real expected
names for this specific dataset, not placeholders.)

```
import pandas as pd
import matplotlib.pyplot as plt
```

**Load**

```
df = pd.read_csv("wages_by_education.csv")
df.head()
```

**Inspect**

```
df.columns
```

**Clean / subset**

```
sub = df[["year", "high_school", "bachelors_degree"]].copy()
sub["year"] = sub["year"].astype(int)
sub["high_school"] = pd.to_numeric(sub["high_school"], errors="coerce")
sub["bachelors_degree"] = pd.to_numeric(sub["bachelors_degree"], errors="coerce")
sub = sub.dropna(subset=["high_school", "bachelors_degree"])
```

**Compute**

```
sub["ba_premium_ratio"] = sub["bachelors_degree"] / sub["high_school"]
sub["ba_premium_dollars"] = sub["bachelors_degree"] - sub["high_school"]
sub.tail()
```

**Visualize**

```
ax = sub.plot(x="year", y="ba_premium_ratio",
              title="BA / HS Wage Premium Over Time", legend=False)
ax.set_ylabel("BA wage divided by HS wage")
plt.show()
```

**Summarize**

```
first5 = sub.head(5).mean(numeric_only=True)
last5 = sub.tail(5).mean(numeric_only=True)

summary = {
    "avg_year_first5": float(first5["year"]),
    "avg_year_last5": float(last5["year"]),
    "ratio_first5": float(first5["ba_premium_ratio"]),
    "ratio_last5": float(last5["ba_premium_ratio"]),
    "dollars_first5": float(first5["ba_premium_dollars"]),
    "dollars_last5": float(last5["ba_premium_dollars"]),
}
summary
```
