Data Visualization — Python
Level: Beginner-Intermediate | Prerequisites: 01-data-cleaning, 02-data-analysis

Copy each block into its own Jupyter cell. Assumes a cleaned DataFrame `sub`.

NOTE ON PLACEHOLDERS: "column_a" and "column_b" below are PLACEHOLDER
NAMES standing in for your actual column names. Replace them before running.

```
import matplotlib.pyplot as plt

# Optional: nicer default figure size
plt.rcParams["figure.figsize"] = (9, 5)
```

**Line chart: trend over time**

```
sub.plot(x="year", y=["column_a", "column_b"])  # <-- PLACEHOLDER columns
plt.title("Column A and B Over Time")
plt.ylabel("Value")
plt.show()
```

**Scatterplot: relationship between two variables**

```
sub.plot.scatter(x="column_a", y="column_b")  # <-- PLACEHOLDER columns
plt.title("Column A vs Column B")
plt.show()
```

**Bar chart: comparing a few categories/years**

```
few = sub[sub["year"].isin([2000, 2010, 2020])]
few.set_index("year")[["column_a", "column_b"]].plot(kind="bar")  # <-- PLACEHOLDER columns
plt.title("Values by Selected Years")
plt.ylabel("Value")
plt.show()
```

**Histogram: distribution of one variable**

```
sub["ratio"].plot(kind="hist", bins=20)
plt.title("Distribution of Ratio")
plt.xlabel("Ratio")
plt.show()
```

**Boxplot: compare distributions across variables**

```
sub[["column_a", "column_b"]].plot(kind="box")  # <-- PLACEHOLDER columns
plt.title("Distribution Comparison")
plt.show()
```
