**Computing a derived variable (ratio + difference)**

```
# PLACEHOLDER columns "column_a"/"column_b" — replace with your real columns
sub["ratio"] = sub["column_a"] / sub["column_b"]  # <-- PLACEHOLDER
sub["difference"] = sub["column_a"] - sub["column_b"]  # <-- PLACEHOLDER
sub.tail()
```

**Summarizing change over time: early period vs. late period**

```
first5 = sub.head(5).mean(numeric_only=True)
last5 = sub.tail(5).mean(numeric_only=True)

summary = {
    "avg_year_first5": float(first5["year"]),
    "avg_year_last5": float(last5["year"]),
    "ratio_first5": float(first5["ratio"]),
    "ratio_last5": float(last5["ratio"]),
    "difference_first5": float(first5["difference"]),
    "difference_last5": float(last5["difference"]),
}
summary
```
