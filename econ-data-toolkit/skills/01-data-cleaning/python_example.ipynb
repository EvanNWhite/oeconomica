Data Cleaning — Python
Level: Beginner | Prerequisites: none

This is written as a plain script but every block is meant to be run as
its own cell in a Jupyter notebook. Copy each block into a new cell.

NOTE ON PLACEHOLDERS: "your_data.csv", "column_a", "column_b" below are
PLACEHOLDER NAMES — they do not refer to a real file or real columns.
Replace them with your actual filename and column names before running.

**Cell 1: imports**

```
import pandas as pd
```

**Cell 2: load the data**

```
# PLACEHOLDER: replace "your_data.csv" with the path to your own CSV.
# Easiest option: put the CSV in the same folder as your notebook and just
# use the filename.
path = "your_data.csv"  # <-- PLACEHOLDER, replace with your real file
df = pd.read_csv(path)
df.head()  # always look at your data before doing anything to it
```

**Cell 3: inspect the columns**

```
df.columns
# Look for the columns relevant to your research question. Note the exact
# spelling/capitalization — you'll need it in the next step.
```

**Cell 4: subset to only what you need**

```
# PLACEHOLDER: "column_a" and "column_b" are stand-ins — replace with your
# actual column names from df.columns above.
sub = df[["year", "column_a", "column_b"]].copy()  # <-- PLACEHOLDER columns
sub.head()

# Why .copy()? Without it, `sub` can act as a "view" into `df` rather than
# an independent object. If you later edit `sub`, you risk unexpected
# changes propagating back. .copy() guarantees a fully separate object.
```

**Cell 5: check for missing values**

```
sub.isna().sum()
```

**Cell 6: fix types**

```
# CSVs sometimes load numeric columns as text (especially if there are
# stray symbols like "$" or "," in the original file).
sub["year"] = sub["year"].astype(int)
sub["column_a"] = pd.to_numeric(sub["column_a"], errors="coerce")  # <-- PLACEHOLDER
sub["column_b"] = pd.to_numeric(sub["column_b"], errors="coerce")  # <-- PLACEHOLDER
# errors="coerce" turns anything that can't be converted into NaN instead
# of crashing — useful for messy real-world data.
```

**Cell 7: drop rows with missing key values (only if needed)**

```
sub = sub.dropna(subset=["column_a", "column_b"])  # <-- PLACEHOLDER columns
```

**Cell 8: sanity check the result**

```
sub.head()
sub.describe()  # quick summary stats to make sure numbers look reasonable
```
